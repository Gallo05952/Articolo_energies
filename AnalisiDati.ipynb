```python
import pandas as pd
import numpy as np
import seaborn as sns
import json
import matplotlib.pyplot as plt
```

```python
#file=["Dati ComposizioneT1.json","Dati ComposizioneT2.json","Dati ComposizioneT3.json","Dati ComposizioneT4.json"]
file=["Dati Composizione3.1.json","Dati Composizione3.2.json","Dati Composizione3.3.json","Dati Composizione3.4.json"]
data = []
df=[]
i=0
for i in range(len(file)):
    with open(file[i], 'r') as f:
        for line in f:
            data.append(json.loads(line))

    df.append(pd.json_normalize(data))
```

```python
copiaLavoro=df.copy()
```

```python
copiaLavoro[0]
```

```python
copiaPulit1=copiaLavoro[0][copiaLavoro[0]["Q_Condenser"].notna()]
copiaPulit2=copiaLavoro[1][copiaLavoro[1]["Q_Condenser"].notna()]
copiaPulit3=copiaLavoro[2][copiaLavoro[2]["Q_Condenser"].notna()]
copiaPulit4=copiaLavoro[3][copiaLavoro[3]["Q_Condenser"].notna()]
copiaPulit=[copiaPulit1,copiaPulit2,copiaPulit3,copiaPulit4]
```

```python
copiaPulit = []

for df in copiaLavoro:
    copiaPulit.append(df[df["Q_Condenser"].notna()])
```

```python
# Create a new figure
plt.figure()

# Define a list of markers
markers = ['o', 'v', '^', '<', '>', 's', 'p', '*', 'X', 'D', 'h', 'H']

# Assuming copiaPulit is a list of DataFrames
for i, df in enumerate(copiaPulit):
    # Add a scatter plot to the figure for each DataFrame
    sns.scatterplot(x="Numero di Sradi", y="Composizioni Finali.CH4", data=df, label=f'DataFrame {i+1}', marker=markers[i % len(markers)])

# Show the plot
plt.show()
```
