```python
import pandas as pd
import numpy as np
import seaborn as sns
import json
import matplotlib.pyplot as plt
```

```python
#file=["Dati ComposizioneT1.json","Dati ComposizioneT2.json","Dati ComposizioneT3.json","Dati ComposizioneT4.json"]
file=["Dati/Dati Composizione3.1.json","Dati/Dati Composizione3.2.json","Dati/Dati Composizione3.3.json","Dati/Dati Composizione3.4.json"]
data = []
df=[]
i=0
for i in range(len(file)):
    with open(file[i], 'r') as f:
        for line in f:
            data.append(json.loads(line))

    df.append(pd.json_normalize(data))
```
